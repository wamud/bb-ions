```python
import stim
import sys
import os
sys.path.append(os.path.abspath("../src"))
from bb_ions import *


# # [[30, 4, 5]] BB5 (weight-5 stabilisers) code from Ye Delfosse long chain [2503.22071] Table II
# l = 5
# m = 3
# # A = x^0 + x
# # B = x^0 + y + x^2*y^2
# Aij = [(0, 0), (1, 0)]          # the powers (i, j) of each term x^i * y^j in A
# Bij = [(0, 0), (0, 1), (2, 2)]
# d = 5

# [[144, 12, 12]] BB6 'gross' code from BB paper [2308.07915] Table III
l = 12
m = 6
# A = x^3 + y + y^2
# B = y^3 + x + x^2
Aij = [(3, 0), (0, 1), (0, 2)]
Bij = [(0, 3), (1, 0), (2, 0)]
d = 12


## Options for memory experiment circuit:

memory_basis = 'Z' 
num_syndrome_extraction_cycles = d
ps = [0.001, 0.002] # physical error rate values



## Make circuits: 

code = get_code_params(l, m, Aij, Bij, d)

# Clear last example circuits:
empty_folder('example_circuits')
empty_folder('example_circuit_diagrams')


# Generate circuits:
for p in ps:


    # Set all errors to zero before manually setting them:
    errors = zero_errors()

    # Manually set errors: (first argument is the error operation that will be applied after (or before in the case of measurement) a specific operation, second is its probability):
    errors["RZ"] = Error("DEPOLARIZE1", p ) # note reset error is currently DEPOLARIZE1 (as per longchain [2503.22071] paper) as opposed to orthogonal-basis state
    errors["RX"] = Error("DEPOLARIZE1", p ) #  "
    errors["H"] = Error("DEPOLARIZE1", p )
    errors["CNOT"] = Error("DEPOLARIZE2", p)
    errors["CZ"] = Error("DEPOLARIZE2", p)
    errors["MZ"] = Error("X_ERROR", p )
    errors["MX"] = Error("Z_ERROR", p )

    # Shuttling errors:
    
    errors["shift_constant"] = p / 10 # determines what the cyclic shift shuttling error rate is proportional to
    errors["shift"] = Error("DEPOLARIZE1", None) # A shuttling error rate associated with cyclically shifting check qubit modules around the racetrack. The p value will be updated in the make_circuit function using p = shift_constant * dist, where dist is the number of data qubit modules that a check qubit module needs to be shuttled past before arriving at the one it will interact with    
    



    # Idling errors - the errors that occur and with what probabiliy while specific operations are being performed on other qubits and this qubit sits idle:
    idle_during = zero_idling(p)

    idle_during["RZ"] = Error("DEPOLARIZE1", p ) 
    idle_during["RX"] = Error("DEPOLARIZE1", p ) 
    idle_during["H"] = Error("DEPOLARIZE1", p )
    idle_during["CNOT"] = Error("DEPOLARIZE1", p)
    idle_during["CZ"] = Error("DEPOLARIZE1", p)
    idle_during["MZ"] = Error("DEPOLARIZE1", p )
    idle_during["MX"] = Error("DEPOLARIZE1", p )

    # Idle while other qubits are shuttling:
    idle_during["shift_constant"] = p / 10
    idle_during["shift"] = Error("DEPOLARIZE1", None) 



    # Make circuit:

    circuit = make_circuit(  # (see src/bb_ions/circfuncs for explanation of make_circuit inputs)
        code,  
        memory_basis,  
        p,  
        num_syndrome_extraction_cycles,  
        errors,
        idle_during,
        sequential_gates = False, 
        exclude_opposite_basis_detectors = True,
        reuse_check_qubits = True,  
    )

    # Save circuit and diagram:
    filename = f"nkd=[[{code.n}_{code.k}_{code.d_max}]],p={p},b={memory_basis},r={num_syndrome_extraction_cycles},code=BB,l={l},m={m},A='{''.join(str(x) + str(y) for x, y in Aij)}',B='{''.join(str(x) + str(y) for x, y in Bij)}'"
    circuit.to_file(f"example_circuits/{filename}.stim")
    svg = str(circuit.diagram("timeline-svg"))
    with open(f"example_circuit_diagrams/{filename}.svg", "w", encoding="utf-8") as f: f.write(svg)
```

```python
# # Run memory experiment simulations:

import sinter
import glob
from stimbposd import SinterDecoder_BPOSD, sinter_decoders

circuit_paths = glob.glob(f"example_circuits/nkd=[[{code.n}_{code.k}_{code.d_max}*.stim")

csv_path = "example_stats/collected_stats.csv"

tasks = [
    sinter.Task(
        circuit_path = path,
        json_metadata = sinter.comma_separated_key_values(path),
    )
    for path in circuit_paths
]


empty_folder('example_stats') # clear previous example stats

# Collect new stats:
samples = sinter.collect(
    num_workers=4,
    max_shots=1000,
    max_errors=1000,
    tasks = tasks,
    decoders=['bposd'],
    save_resume_filepath = csv_path,
    custom_decoders = sinter_decoders()
    )
```

```python
# Plot results:

import matplotlib.pyplot as plt

# Plot:
collected_stats = sinter.stats_from_csv_files(f'example_stats/collected_stats.csv')

fig, ax = plt.subplots(1, 1)

sinter.plot_error_rate(
    ax=ax,
    stats=collected_stats,
    x_func=lambda stats: stats.json_metadata['p'],
    group_func=lambda stats: stats.json_metadata['nkd'],
    failure_units_per_shot_func = lambda stats: stats.json_metadata['r'], # set as num_rounds ('r') failure units to find pL per round
)

ax.loglog()
ax.legend()
ax.set_title("$p_L$ vs. $p$")
ax.set_xlabel("Physical Error Rate")
ax.set_ylabel("Logical Error Rate per round");
```
